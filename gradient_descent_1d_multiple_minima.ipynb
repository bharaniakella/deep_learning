```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
def fx(x):
  return np.cos(2*np.pi*x)+x**2
```

```python
def deriv(x):
  return -np.sin(2*np.pi*x)*2*np.pi + 2*x
```

```python
x=np.linspace(-2,2,2001)
#graph for fx
plt.plot(x,fx(x))
plt.show()
```

```python
#graph for derivative of x
plt.plot(x,deriv(x))
plt.show()
```

```python
plt.plot(x,fx(x),x,deriv(x))
plt.legend(["fx","deriv x"])
plt.show()
```

```python

localmin=np.random.choice(x,1)
print("Initial localmin: ",localmin)
epochs=100
learning_rate=0.01
for i in range(epochs):
  grad=deriv(localmin)
  localmin=localmin-learning_rate*grad

print("localmin after gradient descent: ",localmin)
```

```python
localmin=np.random.choice(x,1)
param=np.zeros((epochs,2))
for i in range(epochs):
  grad=deriv(localmin)
  localmin=localmin-learning_rate*grad
  param[i,0]=localmin
  param[i,1]=grad
```

```python
#graph for local min
for i in range(2):
  plt.plot(param[:,0])
```

```python
#graph for grad
for i in range(2):
  plt.plot(param[:,1])
```
