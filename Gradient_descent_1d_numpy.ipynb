```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def fx(x):
  return 3*x**2 - 3*x + 4
```

```python
def deriv(x):
  return 6*x -3
```

```python
x=np.linspace(-2,2,2001)

#plot between x and fx
plt.plot(x,fx(x))
plt.show()
```

```python
#plot between x and deriv x
plt.plot(x,deriv(x))
plt.show()
```

```python
#plot between function and derivative
plt.plot(x,fx(x),x,deriv(x))
plt.show()
```

```python
localmin=np.random.choice(x,1)
localmin
print("Initial localmin: ",localmin)
epochs=100
learning_rate=0.01

#implementing gradient descent algorithm

for i in range(epochs):
  grad=deriv(localmin)
  localmin=localmin-learning_rate*grad

print("localmin after gradient descent: ",localmin)
```

```python
#plotting graph for localmin and grad
localmin=np.random.choice(x,1)
param=np.zeros((epochs,2))
for i in range(epochs):
  grad=deriv(localmin)
  localmin=localmin-learning_rate*grad
  param[i,0]=localmin
  param[i,1]=grad
```

```python
#graph for local min
for i in range(2):
  plt.plot(param[:,0])
```
