```python
#logarithms are monotonic functions
#monotonic functions help in providing distance between really small values
#log is inverse of exponent
```

```python
import numpy as np
a=np.array([1,2,3])
my_log=np.log(a)
my_exp=np.exp(a)
```

```python
print(my_log)
print(my_exp)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
plt.plot(a,my_log)
```

```python
plt.plot(a,my_exp)
```

```python
plt.plot(np.log(my_exp))
```

```python
plt.plot(np.exp(my_log))
```

```python
import torch

torch.log(torch.Tensor(a))
```
